Use the `complete-shakespeare.txt` dataset for this task. Write **regular expressions that provides**:
- all stage instructions (non-dialogue)
- everything spoken by the character `CALIBAN`
- all sentences that end in exclamation points
- a dictionary with the key being the scene name, and the value being a list of all the characters who appear in them.

You can assume that a sentence or dialogue is only ONE line long (ie. I will not take off points if you cannot capture multiline dialogue).

```python
import re
text_file = open('complete-shakespeare.txt').read()
```

##### All stage instructions (non-dialogue)

```python
# extract instructions in []
q_bracket = r'\[\s?(.+)\]'
inst_list_brac = re.findall(q_bracket, text_file.replace('_', ''))
inst_list_brac
```

```python
# reference for stage directions: 
# https://www.nosweatshakespeare.com/resources/shakespeare-stage-directions/
# (and 1 typo "Exuent" in the script)
stg_dir = '(Alarum|Aside|Enter|Epilogue|Excursion|Exeunt|Exuent|Exit|Re-exit|Flourish|Hautboys|Prologue|Sennet|Solus|Torches)'
q_stg_dir = '.*(' + stg_dir + '.*\n?.*)'
inst_tup_l = re.findall(q_stg_dir, text_file)
inst_list_stg_dir = []
for inst in inst_tup_l:
    if ']' in inst[0]: # exclude instructions in brackets
        continue
    inst_list_stg_dir.append(re.sub('[\[\]_\n]', '', re.sub('\n\s*[A-Z]+\..*', '', re.sub(' +', ' ', inst[0]))))
inst_list_stg_dir
```

##### Everything spoken by the character `CALIBAN`

```python
re.findall(r'CALIBAN\.\s*(.+?[\.!?])\s*[A-Z]+\.', text_file.replace('\n', ''))
```

##### All sentences that end in exclamation points

```python
re.findall(r'[A-Z][\w\d\s,;]*!', text_file.replace('\n', ' '))
```

##### A dictionary with the key being the scene name, and the value being a list of all the characters who appear in them.

```python
scene_key = re.findall(r'SCENE.*', text_file)
scene_txt = re.split(r'SCENE.*', text_file)[1:]
scene_dict = {}
for i, scene in enumerate(scene_key):
    scene_dict[str(i+1) + scene_key[i]] = scene_txt[i]
```

```python
for scene in scene_dict:
    char = set(re.findall(r'([A-Z]{4,} ?[A-Z]* ?[A-Z]*)\.', scene_dict[scene]))
    if len(char) == 0:
        char = None
    scene_dict[scene] = char
scene_dict
```
